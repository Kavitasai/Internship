```python
driver = webdriver.Chrome()
```

```python
driver.get('https://www.flipkart.com/')
```

```python
item = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/header/div[1]/div[2]/form/div/div/input')
                           
item.send_keys('sunglasses')
```

```python
search = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/header/div[1]/div[2]/form/div/button')
search.click()
```

```python
product = []
brand =[]
price =[]
discount_offered =[]
```

```python
start = 0
end = 3
max_items = 100
brand = []
product = []
discount_offered = []
price = []

total_items = 0  
for page in range(start, end):
    if total_items >= max_items:
        break  

    
    prod = driver.find_elements(By.XPATH, '//a[@class="IRpwTa"]')
    bran = driver.find_elements(By.XPATH, '//div[@class="_2WkVRV"]')
    price_tags = driver.find_elements(By.XPATH, '//div[@class="_30jeq3"]')
    discount = driver.find_elements(By.XPATH, '//div[@class="_3Ay6Sb"]')

    items = min(max_items - total_items, len(prod), len(bran), len(price_tags), len(discount))

    for i in range(items):
        product.append(prod[i].text)
        brand.append(bran[i].text)
        price.append(price_tags[i].text)
        discount_offered.append(discount[i].text)

        total_items += 1  

    if total_items >= max_items:
        break

    next_button = driver.find_element(By.XPATH, '//a[@class="_1LKTO3"]')
    next_button.click()
    time.sleep(3)

print(len(brand), len(product), len(discount_offered), len(price))
```

```python
df = pd.DataFrame({'Brand':brand,'Product':product,'Discount_Offered':discount_offered,'Price':price})
df
```

Q5: Scrape 100 reviews data from flipkart.com for iphone11 phone. You have to go the link:

https://www.flipkart.com/apple-iphone-11-black-64-gb/productreviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART

As shown in the above page you have to scrape the tick marked attributes. These are:
1. Rating
2. Review summary
3. Full review
4. You have to scrape this data for first 100reviews


```python
driver = webdriver.Chrome()
```

```python
driver.get('https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&market')
```

```python
rating=[]
review_summary=[]
full_review=[]
```

```python
start = 0
end = 10
max_items = 100  


for page in range(start, end):
    rate_elements = driver.find_elements(By.XPATH, '//div[@class="_3LWZlK _1BLPMq"]')
    review_elements = driver.find_elements(By.XPATH, '//p[@class="_2-N8zT"]')
    full_review_elements = driver.find_elements(By.XPATH, '//div[@class="t-ZTKy"]')

    for i in range(len(rate_elements)):
        rat = rate_elements[i].text
        rating.append(rat)
        
        if i < len(review_elements):
            rew = review_elements[i].text
            review_summary.append(rew)
        else:
            review_summary.append("Data not available")

        if i < len(full_review_elements):
            rev = full_review_elements[i].text
            full_review.append(rev)
        else:
            full_review.append("Data not available")

        if len(rating) >= max_items:
            break

    if len(rating) >= max_items:
        break
    
    next_button = driver.find_element(By.XPATH, '//a[@class="_1LKTO3"]')
    next_button.click()
    time.sleep(3)


while len(rating) < max_items:
    rating.append("Data not available")

while len(review_summary) < max_items:
    review_summary.append("Data not available")

while len(full_review) < max_items:
    full_review.append("Data not available")

print(len(rating), len(review_summary), len(full_review))
```

```python
df = pd.DataFrame({'Rating':rating,'Review_Summary':review_summary,'Full_Review':full_review})
df
```

Q8: Write a python program to scrape data for Top 1000 Quotes of All Time.

The above task will be done in following steps:
1. First get the webpagehttps://www.azquotes.com/
2. Click on TopQuotes
3. Than scrap a) Quote b) Author c) Type Of Quotes

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://www.azquotes.com/')
```

```python
quote = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
quote.click()
```

```python
start = 0
end = 100

quotes = []
authors = []
quote_types = []

for page in range(start, end):
    
    quote_elements = driver.find_elements(By.XPATH, '//a[@class="title"]')
    
    author_elements = driver.find_elements(By.XPATH, '//div[@class="author"]')
    
    quote_type_elements = driver.find_elements(By.XPATH, '//div[@class="tags"]')
    
    for i in range(len(quote_elements)):
        quote = quote_elements[i].text
      
        author = author_elements[i].text
        
        quote_type = quote_type_elements[i].text
        
        quotes.append(quote)
        authors.append(author)
        quote_types.append(quote_type)
    
    try:
        next_button = driver.find_element(By.XPATH, '//a[contains(@class,"next")]/a')
        next_button.click()
    except NoSuchElementException:
       
        break
    
    time.sleep(3)

for i in range(len(quotes)):
    print("Quote:", quotes[i])
    print("Author:", authors[i])
    print("Type of Quote:", quote_types[i])
    print()
```

```python
driver.get('https://www.jagranjosh.com/')
```

```python
gk = driver.find_element(By.XPATH,'/html/body/div/header/nav/div/div/div[3]/ul/li[3]/a')
gk.click()
```

```python
lists = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/div[2]/div/div[10]/div/div/ul/li[2]/a')
lists.click()
```

```python
data = []
table = driver.find_element(By.CLASS_NAME, "table-box")
rows = table.find_elements(By.TAG_NAME, "tr")

for row in rows[1:]:
    cols = row.find_elements(By.TAG_NAME, "td")
    if len(cols) >= 4:
        name = cols[0].text
        born_dead = cols[1].text
        term_of_office = cols[2].text
        remarks = cols[3].text
        data.append([name, born_dead, term_of_office, remarks])

df = pd.DataFrame(data, columns=["Name", "Born-Dead", "Term of Office", "Remarks"])
print(df)
driver.quit()
```
