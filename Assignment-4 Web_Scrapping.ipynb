1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url
= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: A)
Rank
B) Name
C) Artist
D) Upload date
E) Views 

```python
driver = webdriver.Chrome()
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
rank = []
name = []
artist = []
upload_date = []
views = []

try:
    ranks = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
    for i in ranks[:30]:
        rank.append(i.text)
        
except Exception as e:
    print("Error Occured : ",str(e))

try:
    names = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')
    for i in names[:30]:
        name.append(i.text)
        
except Exception as e:
    print("Error Occured : ",str(e))
    
try:
    artists = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')
    for i in artists[:30]:
        artist.append(i.text)
        
except Exception as e:
    print("Error Occured : ",str(e))
    
try:
    view = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
    for i in view[:30]:
        views.append(i.text)
        
except Exception as e:
    print("Error Occured : ",str(e))
    
try:
    uploads = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
    for i in uploads[:30]:
        upload_date.append(i.text)
        
except Exception as e:
    print("Error Occured : ",str(e))
```

```python
print(len(rank),len(upload_date),len(name),len(artist),len(views))
```

```python
df = pd.DataFrame({
    "Rank" : rank,
    "Name" : name,
    "Artist" : artist,
    "Views" : views,
    "Publication Date" : upload_date
})
df
```

E) No of instances

G) Year

Note: - from the home page you have to go to the Show All Dataset page through code. 

```python
driver = webdriver.Chrome()
driver.get('https://archive.ics.uci.edu/')
```

```python
web = driver.find_element(By.XPATH,'//a[@class="btn-primary btn"]')
web.click()
```

```python
exp = driver.find_element(By.XPATH,'//span[@class="swap-on text-primary-content"]')
exp.click()
```

```python
product_list = []

for i in range(0,66):
    
    try:
        
        names = driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
        typ = driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[2]/span')
        tasks = driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[1]/span')
        attri = driver.find_elements(By.XPATH,'//div[@class="grid grid-cols-8 overflow-x-auto"]/table/tbody/tr/td[2]')
        instance = driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[3]/span')
        att = driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[4]/span')
        years = driver.find_elements(By.XPATH,'//div[@class="grid grid-cols-8 overflow-x-auto"]/table/tbody/tr/td[3]')
                                   
        for i in range(len(names)):
            product_dict = {
                "Dataset Name": names[i].text if i < len(names) else '-',
                "Data Type" : typ[i].text if i < len(typ) else '-',
                "Task": tasks[i].text if i < len(tasks) else '-',
                "Attibute Type": attri[i].text if i < len(attri) else '-',
                "No Of Instances" : instance[i].text if i < len(instance) else '-',
                "No Of Attributes" : att[i].text if i < len(att) else '-',
                "Years": years[i].text.split('/')[-1] if i < len(years) else '-'

            }
            product_list.append(product_dict)
                                   
        while True:
                                   
            try:
                next_button = driver.find_element(By.XPATH,'//button[@class="btn-primary btn-sm btn"][2]')
                next_button.click()
                time.sleep(5)
            
            except Exception:
                break
                time.sleep(3)
        
        

    except Exception as e:
        print("An error occurred:", str(e))
    
df = pd.DataFrame(product_list) 
print(df)
```
